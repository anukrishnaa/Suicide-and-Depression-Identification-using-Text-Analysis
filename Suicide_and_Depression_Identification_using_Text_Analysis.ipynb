```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
df =pd.read_csv("/content/reddit_depression_suicidewatch.csv")
```

```python
x=df["text"]
```

```python
y=df["label"]
```

```python
x[:10]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3,random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer(stop_words = 'english',lowercase=False)
# fit the vectorizer on the training data
vector.fit(x_train)
vector.vocabulary_
x_transformed = vector.transform(x_train)
x_transformed.toarray()
# for test data
x_test_transformed = vector.transform(x_test)
```

```python
vector.vocabulary_
```

# LOGISTIC **REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model = model.fit(x_transformed,y_train)
```

```python
y_pred = model.predict(x_test_transformed)
print(classification_report(y_test,y_pred))
```

```python
note = ['''I have no idea what to do after this breakup''']
vec = vector.transform(note).toarray()
print('text:', note)
print(str(list(model.predict(vec))[0]).replace('0', 'depression').replace('1', 'SuicideWatch'))
```

```python
note = ['''Mom,Dad I am sorry I did this. BUt I cant live like this anymore''']
vec = vector.transform(note).toarray()
print('text:', note)
print(str(list(model.predict(vec))[0]).replace('0', 'depression').replace('1', 'SuicideWatch'))
```

# RANDOM FOREST CLASSIFIER

```python
model = RandomForestClassifier(n_estimators=200)
model.fit(x_transformed, y_train)
```

```python
y_pred = model.predict(x_test_transformed)
print(classification_report(y_test,y_pred))
```

# DECISION TREE

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_transformed,y_train)
```

```python
y_pred = dtree.predict(x_test_transformed)
```

```python
print(classification_report(y_test,y_pred))
print("------------")
print(confusion_matrix(y_test,y_pred))
```
